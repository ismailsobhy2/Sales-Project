# Visulization

```python
data_2 = sales_df["Site"].value_counts() ## convert object to numeric
# Create a pie chart
plt.pie(data_2.values, labels=data_2.index, autopct="%1.1f%%")
plt.title("Project Sites")
```

```python
data_22 = sales_df["Severity"].value_counts()
# Define custom colors
custom_colors = ["green", "red", "yellow", "orange"]  # Example colors

# Create a pie chart with custom colors
plt.pie(data_22.values, labels=data_22.index, autopct="%1.1f%%", colors=custom_colors, startangle=90)

# Add title
plt.title("Projects Severity")
# Show the chart
plt.show()
```

```python
data_3 = sales_df["Project_Status"].value_counts()

# Create a bar chart
plt.bar(data_3.index, data_3.values, color="green", edgecolor="black")

# Add labels and title
plt.ylabel("Count")
plt.title("Projects Status")
plt.xticks(rotation=45)  # Rotate labels for better visibility
# Show the chart
plt.show()
```

```python

# Count occurrences of each unique 'Ticket Origin'
data_4 = sales_df["Ticket Origin"].value_counts()

# Generate an alternative visual using a bar chart
plt.figure(figsize=(10, 6))
plt.bar(data_4.index, data_4.values, color='blue', alpha=0.7)
plt.xlabel("Ticket Origin")
plt.ylabel("Count")
plt.title("Ticket Origin Distribution (Bar Chart)")
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.grid(axis='y')

# Show the bar chart
plt.show()

```

```python
# Convert 'Week' column to datetime format
sales_df['Week'] = pd.to_datetime(sales_df['Week'] + '-1', format='%Y_W%W-%w', errors='coerce')

# Aggregate data by week
weekly_trends = sales_df.groupby('Week')[['Hours', 'Forecasted hours', 'Budget Hours']].sum().reset_index()
# Plot trend analysis
plt.figure(figsize=(12, 6))
plt.plot(weekly_trends['Week'], weekly_trends['Hours'], label='Actual Hours', marker='o')
plt.plot(weekly_trends['Week'], weekly_trends['Forecasted hours'], label='Forecasted Hours', marker='s')
plt.plot(weekly_trends['Week'], weekly_trends['Budget Hours'], label='Budgeted Hours', marker='^')

plt.xlabel('Week')
plt.ylabel('Hours')
plt.title('Trend of Hours, Forecasted Hours, and Budget Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Average cost rate per project status
avg_cost_status = sales_df.groupby('Project_Status')['Cost Rate'].mean().sort_values()

plt.figure(figsize=(10, 6))
plt.bar(avg_cost_status.index, avg_cost_status.values, color='purple')
plt.xlabel('Project Status')
plt.ylabel('Average Cost Rate')
plt.title('Average Cost Rate by Project Status')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Department-wise project distribution
department_counts = sales_df['Department'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(department_counts.index, department_counts.values, color='royalblue')
plt.xlabel('Department')
plt.ylabel('Number of Projects')
plt.title('Project Distribution by Department')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Ticket Origin Distribution by Severity
severity_origin = sales_df.pivot_table(index='Ticket Origin', columns='Severity', aggfunc='size', fill_value=0)

severity_origin.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')
plt.xlabel('Ticket Origin')
plt.ylabel('Number of Tickets')
plt.title('Ticket Origin Distribution by Severity')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Severity')
plt.show()
```
