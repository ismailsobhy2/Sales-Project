# Inner Join two dataframe

```python
sales_df = pd.merge(Hours_df, Eng_df, on="Project Code", how="inner")  
sales_df
```

```python
sales_df.to_csv("sales_cleaned.csv", index=True)
```

# Visulization

```python
data_1 = sales_df["Department"].value_counts()

# Create a bar chart
plt.bar(data_1.index, data_1.values, color="green", edgecolor="black")

# Add labels and title
plt.ylabel("Count")
plt.title("Department Distribution")
plt.xticks(rotation=45)  # Rotate labels for better visibility
# Show the chart
plt.show()
```

```python
data_2 = sales_df["Site"].value_counts() ## convert object to numeric
# Create a pie chart
plt.pie(data_2.values, labels=data_2.index, autopct="%1.1f%%")
plt.title("Project Sites")
```

```python
data_22 = sales_df["Severity"].value_counts()
# Define custom colors
custom_colors = ["green", "red", "yellow", "orange"]  # Example colors

# Create a pie chart with custom colors
plt.pie(data_22.values, labels=data_22.index, autopct="%1.1f%%", colors=custom_colors, startangle=90)

# Add title
plt.title("Projects Severity")
# Show the chart
plt.show()
```

```python
data_3 = sales_df["Project_Status"].value_counts()

# Create a bar chart
plt.bar(data_3.index, data_3.values, color="green", edgecolor="black")

# Add labels and title
plt.ylabel("Count")
plt.title("Projects Status")
plt.xticks(rotation=45)  # Rotate labels for better visibility
# Show the chart
plt.show()
```

```python
# Count occurrences of each unique 'Ticket Origin'
data_4 = sales_df["Ticket Origin"].value_counts()

# Create a line graph
plt.plot(data_4.index, data_4.values, marker='o', linestyle='-', color='blue')

# Add labels and title
plt.ylabel("Count")
plt.title("Ticket Origin Distribution")
plt.xticks(rotation=45)  # Rotate labels for better visibility

# Show the graph
plt.show()
```
